## Cross-validation of time-series forecaster

With time-series forecasting we have to be careful about how we cross-validate our models.
Our data has a temporal ordering and we must use the *past* to predict the *future*.

The two main choices are
* **Sliding window cross-validation**: We take a fixed window size and *train* our forecaster on this data. After this we make predictions and compare them against the *forecasting* window. When then shift this window along. This keeps the amount of data used to fit the models fixed in each of the folds.

* **Expanding window cross-validation**: The window used to train our models has a fixed start and we keep increasing its size. The amount of data used to fit models increases with each of the folds.

I find the *sliding window* method more attractive method as each window contains the same amount of data.

Taken from [Forecasting at Uber: An Introduction](https://www.uber.com/en-GB/blog/forecasting-introduction/)


## Baseline forecasters

To establish whether a forecasting method we need a baseline prediction. For classification problems we can establish a baseline by predicting the dominate class. For regression problems we can predict the mean as a baseline.

For forecasting problems we can establish a naive baseline by predicting the previous value in the series. If we have some knowledge about the periodicity of the time-series it makes sense to predict the value from the previous *period*, for example predict the previous days/weeks value.

Lets do this and see what comes out.

First we need a Sliding Window Splitter to perform the cross-validation.
Lets take 180 days worth of data in the window, this is the window which contains the data we will fit the data on.
We will step the window forward a week plus two hours each time. The two hours is there to make sure we don't start the forecasting from the same day of the week/hour each time.

We also need a forecasting horizon. Lets make predictions up to a month in advance

```python
fh = np.arange(1,4*7*24)
cv = SlidingWindowSplitter(
        fh=fh, 
        window_length=180*24, #Take half a years worth of data
        step_length = 7*24 + 2 #Slide the window forward one week and two hours.
    )
```

Lets setup naive forecasters that look back one hour, one day and one week and put them in a dictionary

```python
naive_lasthour = NaiveForecaster(strategy='last',sp=1)
naive_lastday  = NaiveForecaster(strategy='last',sp=1*24)
naive_lastweek = NaiveForecaster(strategy='last',sp=7*24)

forecasters = {
              "Naive Last Hour":naive_lasthour,
              "Naive Last Day":naive_lastday,
              "Naive Last Week":naive_lastweek,
             }
```

We also need a way of scoring these
Here I choose symmetric MeanAbsolutePercentageError

```python
mape_sym = MeanAbsolutePercentageError(symmetric=True)
```

Now I am going to evaluate these models. This means for each of the windows from the Sliding Window Splitter I will use the forecaster to make a predictions.

This might take a moment.

```python
evals = {forecaster_name:
             {col:evaluate(forecaster=forecaster,
                        y=df_wide[col],
                        cv=cv,
                        strategy="refit",
                        return_data=True,
                        scoring = [mape_sym]
                        ) 
              for col in tqdm.tqdm(df_wide.columns,desc=f"{forecaster_name}") }
                  for forecaster_name,forecaster in forecasters.items()}
```

Great now we have a dictionary of dataframes containing details of the evaluations in the sliding window. Including the data used to train and test and the prediction made by the forecaser

```python
evals['Naive Last Hour']['New Quay'].head()
```

Lets plot some off these and take a look

```python
def plotCVSeries(evals,i,ax,tail=24*7,head=24*7):
    plot_series(evals['y_train'].iloc[i].tail(tail),
                evals['y_test'].iloc[i].head(head),
                evals['y_pred'].iloc[i].head(head),
                labels=['Train','Test','Pred'],
                ax = ax)
```

```python
fig,axs = plt.subplots(3,1,figsize=(12,9))

location = locations[5]
for forecaster_name,ax in zip( forecasters.keys(),axs ):
    plotCVSeries(evals[forecaster_name][location],0,ax)
    ax.set_title(f"{forecaster_name}")
    
for ax in axs:
    ax.set_xticklabels([])
plt.tight_layout()
```

Unfortunately the scores returned are only the evaluated over the entire horizon we asked for.
I want to see how the error changes when the horizon changes.

I have some routines ready to do this.

```python
df_scores = getScoreTable(evals)
```

We now have the Score for each fold, horizon and location

### Linear regression forecaster.

We can apply any regression technique we can think of at a forecasting problem.
So lets start with linear regression.

```python
linear_forecaster = lambda window_length: make_reduction(LinearRegression(),
                                                         window_length=window_length,
                                                         strategy='recursive')

forecasters = {
              "Linear 12 hours":linear_forecaster(12),
              "Linear 24 hours":linear_forecaster(24),
              "Linear 72 hours":linear_forecaster(72),
              "Naive Last Week":NaiveForecaster(strategy='last',sp=7*24),
             }

evals = {forecaster_name:
             {col:evaluate(forecaster=forecaster,
                        y=df_wide[col],
                        cv=cv,
                        strategy="refit",
                        return_data=True,
                        scoring = [mape_sym]
                        ) 
              for col in tqdm.tqdm(df_wide.columns,desc=f"{forecaster_name}") }
                  for forecaster_name,forecaster in forecasters.items()}
```

```python
df_scores = getScoreTable(evals)

g = sns.FacetGrid(df_scores, col='Location', col_wrap=5)
g.map_dataframe(sns.lineplot, x='Horizon', y='Score', hue='Forecaster', errorbar="sd")
g.set(xticks=np.arange(0, 7*4+1))
g.set_titles("{col_name}")
g.add_legend()
```
